```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi=[10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi=[20,60,80,85,86,87,70,90,95,80]
```

```python
numpyYasListesi=np.array(yasListesi)
numpyKiloListesi=np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"b")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Grafiği")
#plt.show()
```

```python
numpydizisi1=np.linspace(0,10,20)
```

```python
numpydizisi1
```

```python
numpydizisi2=numpydizisi1**3
```

```python
numpydizisi2
```

```python
plt.plot(numpydizisi1,numpydizisi2,"b*-")#"b--"
```

```python
plt.subplot(1,2,1)
plt.plot(numpydizisi1,numpydizisi2,"r*-")
plt.subplot(1,2,2)#has 1 line,will be 2 column,right now ı'm drowing 2'nd graph
plt.plot(numpydizisi2,numpydizisi1,"b*-")
```

```python
myFigure=plt.figure()
figureAxes=myFigure.add_axes([0.4,0.6,0.6,0.6])
figureAxes.plot(numpydizisi1,numpydizisi2,"b")
figureAxes.set_xlabel("Xekseni")
figureAxes.set_ylabel("Yekseni")
figureAxes.set_title("Title")
```

```python
figure=plt.figure()
axis1=figure.add_axes([0.1,0.1,0.9,0.9])
axis2=figure.add_axes([0.2,0.5,0.3,0.3])

axis1.plot(numpydizisi1,numpydizisi2,"r")
axis1.set_xlabel("X label")
axis1.set_ylabel("Y label")
axis1.set_title("Main Title")

axis2.plot(numpydizisi2,numpydizisi1,"g")
axis2.set_xlabel("Second X label")
axis2.set_ylabel("Second Y label")
axis2.set_title("Second Title")
```

```python
(myfigure1,myaxis) =plt.subplots(nrows=1,ncols=2)
myaxis.plot(numpydizisi1,numpydizisi2,"r")
```
